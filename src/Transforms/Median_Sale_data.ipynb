```python
import pandas as pd
import numpy as np

zip_median_load = "Sale_Prices_Zip.csv"

sale_price = pd.read_csv(zip_median_load)
sale_price.head()
```

```python
sale_price1 = sale_price.rename(columns={'RegionName':'zip_code'})
```

```python
va_sale_price = sale_price1.loc[sale_price['StateName'] == "Virginia"]
va_sale_price
```

```python
va_2018_price = va_sale_price[['zip_code','StateName', '2018-01', '2018-02','2018-03','2018-04','2018-05','2018-06','2018-07','2018-08','2018-09','2018-10','2018-11','2018-12']]
va_2018_price.head()
```

```python
sale_zip_final_price = va_2018_price[['2018-01', '2018-02','2018-03','2018-04','2018-05','2018-06','2018-07','2018-08','2018-09','2018-10','2018-11','2018-12']].mean(axis=1)
sale_zip_final_price
```

```python
va_2018_price = va_2018_price.assign(sale_zip_final_price=sale_zip_final_price.values)
va_2018_price
```

```python
final =  va_2018_price[['zip_code','StateName','sale_zip_final_price']]
final1 = final.rename(columns={'sale_zip_final_price': '2018_median_sale_price','StateName':'state_name'})
final2 = final1.round(2)
final2
```
